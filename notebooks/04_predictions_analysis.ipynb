```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
# Setup and data loading
plt.style.use('seaborn-v0_8')
sns.set_palette("Set3")
pd.set_option('display.max_columns', None)

predictions_df = pd.read_csv('../data/predictions/test_predictions.csv')


print(f"Training set: {predictions_df.shape}")
print(f"Training set: {predictions_df.info()}")
```

```python
binary_predictions = predictions_df['y'].apply(lambda x: 1 if x >= 0.5 else 0)
count_results = binary_predictions.value_counts()
print(count_results)

plt.figure(figsize=(8, 6))
ax = sns.countplot(x=binary_predictions)
plt.xlabel('Prediction')
plt.ylabel('Count')
plt.title('Test predicted target distribution')

# Add value labels on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d')

plt.savefig('../docs/docs/images/predictions_distribution.png', dpi=300, bbox_inches='tight')

plt.show()
```
